We realised that with some our hypotheses that we need to seperate the car brand from the car model. 

```python
car_brands = [
    'alfa-romero', 'audi', 'bmw', 'chevrolet', 'dodge', 'honda', 'isuzu',
    'jaguar', 'mazda', 'buick', 'mercury', 'mitsubishi', 'nissan', 'peugeot',
    'plymouth', 'porsche', 'renault', 'saab', 'subaru', 'toyota',
    'volkswagen', 'volvo'
]
```

Here we have listed all the car brands.

```python
def extract_brand_model(name, brand_list):
    for brand in brand_list:
        if name.startswith(brand):
            return pd.Series([brand, name[len(brand):].strip()])
    return pd.Series([None, name])  # fallback if no brand match

df[['Brand', 'Model']] = df['CarName'].apply(lambda x: extract_brand_model(x, car_brands))
```

Here we are seperating the car brands from their models.

```python
df = df.drop_duplicates(subset=['Brand', 'Model'])
```

Here the brand and model is seperated.

```python
# Define new column order
new_order = ['Brand', 'Model'] + [col for col in df.columns if col not in ['Brand', 'Model']]

# Reorder the DataFrame
df = df[new_order]
```

Here we have re ordered the dataframe so that the new columns are at the front.

```python
df = df.reset_index(drop=True)
```

Here we have reset the index for dropping duplicates

```python
# Move 'car_ID' to the front
df = df[['car_ID'] + [col for col in df.columns if col != 'car_ID']]
```

Here we have moved the 'car_ID' back to its original position.

```python
df
```

```python
df = df.drop(columns=['CarName'])
```

Finally, we have removed 'CarName' as it is not required anymore due to the seperation of brand and model.
